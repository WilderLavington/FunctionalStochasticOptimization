```python
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
#
z_1 = np.cumsum(np.random.rand(100))
z_2 = np.cumsum(np.random.rand(100))
z_3 = np.cumsum(np.random.rand(100))
#
y_1 = np.random.rand(100) ** 2
y_2 = np.random.rand(100) ** 2.5
y_3 = np.cumsum(np.random.rand(100))
```

```python
f_1 = CubicSpline(z_1, y_1, bc_type='natural')
f_2 = CubicSpline(z_2, y_2, bc_type='naural')
f_3 = CubicSpline(z_3, y_3, bc_type='natural')
x_max = np.max([z_1[-1], z_2[-1], z_3[-1]]) 
x = np.linspace(0,x_max,100)
```

```python
y_1 = f_1(x)
y_2 = f_2(x)
y_3 = f_3(x)
y = (y_1+y_2+y_3)/3
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y , 'r' )
plt.plot(z_1, y_1 , 'b' )
plt.plot(z_2, y_2 , 'b' )
plt.plot(z_3, y_3 , 'b' )
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
